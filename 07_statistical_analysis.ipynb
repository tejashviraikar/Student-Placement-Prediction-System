```python
import pandas as pd
import scipy.stats as stats

# Load dataset
df = pd.read_csv("Job_Placement_Data_Enhanced.csv")

# Select numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

print("Statistical Summary:\n")
print(numeric_df.describe())
```

```python
# Correlation Analysis (Pearson Correlation)
print("\nCorrelation Matrix:\n")
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
# Hypothesis Testing (Pearson Test Between Two Variables)
cols = numeric_df.columns

if len(cols) >= 2:
    col1 = cols[0]
    col2 = cols[1]

    correlation, p_value = stats.pearsonr(numeric_df[col1], numeric_df[col2])

    print(f"\nHypothesis Test Between {col1} and {col2}")
    print("Pearson Correlation:", correlation)
    print("P-value:", p_value)

    if p_value < 0.05:
        print("Result: Statistically significant relationship (Reject Null Hypothesis)")
    else:
        print("Result: Not statistically significant (Fail to Reject Null Hypothesis)")
```
