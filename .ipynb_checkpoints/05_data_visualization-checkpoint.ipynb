```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("Job_Placement_Data_Enhanced.csv")

sns.set_style("whitegrid")
```

```python
# Select numeric columns only
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Histograms for all numeric columns
numeric_df.hist(figsize=(12, 8), bins=20)
plt.suptitle("Distribution of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# Boxplots for Outlier Detection
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_df)
plt.xticks(rotation=45)
plt.title("Boxplot of Numeric Features")
plt.show()
```

```python
# Pairplot (Feature Relationships)
sns.pairplot(numeric_df)
plt.show()
```

```python
# Scatter Plot Between First Two Columns
cols = numeric_df.columns
if len(cols) >= 2:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=numeric_df[cols[0]], y=numeric_df[cols[1]])
    plt.xlabel(cols[0])
    plt.ylabel(cols[1])
    plt.title(f"{cols[0]} vs {cols[1]}")
    plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
